```python
# Get the images' list
result_imgs = os.listdir(save_dir)
result_imgs = map(lambda x: os.path.join(save_dir, x),  result_imgs)
result_imgs = list(result_imgs)

# Display the images
fig = plt.figure(figsize=(14,12))
n = 1
for i in result_imgs[0:n_imgs]:
  img = cv2.imread(i)
  fig.add_subplot(2, 2, n)
  plt.imshow(img)
  n += 1
plt.show()
```
